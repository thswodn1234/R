```python

import pandas as pd
```

```python
df = pd.read_csv("../03/03_암발생자수_.csv",encoding='cp949')
df
```

```python
df.columns
```

```python
df.columns = ['암종별', '성별', '연령별', '2019', '2019.1']
df.columns
```

```python
df22 = df[(df['암종별'] == '모든 암(C00-C96)') & (df['연령별'] !='계') & (df['연령별'] !='연령미상')] 
```

```python
df22
```

```python
import numpy as np
```

```python
df22['연령대'] = np.where(df22['연령별'].isin(["0-4세",  "5-9세","10-14세" , "15-19세"  ,"20-24세" , "25-29세"  ,"30-34세" , "35-39세"]) , "30대 이하",
np.where(df22['연령별'].isin(["40-44세" , "45-49세" , "50-54세",  "55-59세"]), "40대~50대",
np.where(df22['연령별'].isin(["60-64세" , "65-69세",  "70-74세" , "75-79세"]) , "60대~70대","80대이상")))
```

```python
df22['연령대']
```

```python
df22
```

```python
df22g = df22.groupby(['연령대', '성별']).sum()['2019']
df22g
```
